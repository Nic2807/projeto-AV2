### Lista da Fauna e Flora ameaçados de extinção

A tabela contém o anexo da portaria lançada pelo Ministério do Meio Ambiente das espécies da fauna e flora brasileira ameaçadas ou em extinção.

Os dados foram coletados entre 2020 e 2022, pelo Ministério do Meio Ambiente (MMA), criado em novembro de 1992, tem como missão formular e implementar políticas públicas ambientais nacionais de forma articulada e pactuada com os atores públicos e a sociedade para o desenvolvimento sustentável. A visão de futuro do MMA é ser reconhecido pela sociedade e pelo conjunto de atores públicos por sua excelência, credibilidade e eficiência na proteção do meio ambiente.

# Programa

```python
# Import da biblioteca "basedosdados" e renomeação para "bd".

import basedosdados as bd
```

```python
# Para carregar o dado direto no pandas.

df = bd.read_table(dataset_id='br_mma_extincao',
table_id='fauna_ameacada',
billing_project_id="projetoav2-389419")
```

```python
# Retirei uma das colunas do data frame.

df = df.loc[0:1258,('grupo','ordem','familia','especie_ou_subespecie','categoria')]
```

```python
# Variável com as siglas e seus significados para auxiliar o usuário.

siglas = ("VU = Vulnerável\nCR = Perigo Crítico\nPEX = Provavelmente Extinto\nEW = Extinta Na Natureza\nRE = Regionalmente Extinta\nEX = Extinta")
```

```python
# Opções de parâmetros aceitos para filtrar os dados.

filtro = input(str('Você deseja filtrar por:\n Grupo\n Ordem\n Familia\n Categoria\n ')).lower()
```

```python
# Reconhecimento de parâmetro de filtragem.

# Caso o parâmetro escolhido seja "grupo" o programa vai identificar e pedir um grupo específico ao usuário.
if filtro == 'grupo':
    filtro_d = input(str('Insira o grupo que deseja ver:\n')).lower()
    df_result = df.query(f'grupo == "{filtro_d}"')
    
# Caso o parâmetro escolhido seja "ordem" o programa vai identificar e pedir uma ordem específica ao usuário. 
    
if filtro == 'ordem':
    filtro_d = input(str('Insira a ordem que deseja ver:\n')).title()
    df_result = df.query(f'ordem == "{filtro_d}"')
    
# Caso o parâmetro escolhido seja "familia" o programa vai identificar e pedir uma familia específica ao usuário.
    
if filtro == 'familia':
    filtro_d = input(str('Insira a familia que deseja ver:\n')).title()
    df_result = df.query(f'familia == "{filtro_d}"')
    
# Caso o parâmetro escolhido seja "categoria" o programa vai identificar e pedir uma categoria específica ao usuário.
    
if filtro == 'categoria':
    filtro_d = input(str(f'Insira a categoria que deseja ver:\n{siglas}\n')).upper()
    df_result = df.query(f'categoria == "{filtro_d}"')
    
# Em todos os casos o dataframe filtrado se torna a variável "df_resut" que é sempre exibida ao final do código.
    
df_result
```
